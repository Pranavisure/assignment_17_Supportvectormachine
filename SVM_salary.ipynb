```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

```

### Reading train dataset

```python
train_data=pd.read_csv('SalaryData_Train.csv')
train_data
```

#### Data Preprocessing

```python
train_data.columns
```

```python
train_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
train_data.iloc[:, -1] = labelencoder.fit_transform(train_data.iloc[:,-1])
train_data
```

```python
train_data_x=pd.get_dummies(train_data.iloc[:,:13])
train_data_x
```

```python
x=train_data_x
x
```

```python
y=train_data['Salary']
y
```

##### Splitting into train and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

### SVM Model

```python
clf=SVC()
param_grid=[{'kernel':['rbf'],'gamma':[50,5,0.5],'C':[0.1]}]
gsv=GridSearchCV(clf,param_grid)
```

```python
gsv.fit(x_train,y_train)
```

###### Best params

```python
gsv.best_params_ , gsv.best_score_
```

#### Creating model with above params

```python
model_1 = SVC(C= 0.1, gamma = 0.5)
model_1.fit(x_train , y_train)
y_pred = model_1.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

#### Testing accuracy with other values

```python
model_2 = SVC(C= 0.01, gamma = 0.5,kernel='rbf')
model_2.fit(x_train , y_train)
y_pred = model_2.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

#### Model with gamma default

```python
model_final=SVC(C=0.01,kernel='rbf')
model_final.fit(x_train,y_train)
```

#### Making predictions for the above model

```python
y_pred=model_final.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy=",acc)
```
